```python
import pandas as pd
```

```python
# The path to our CSV file
kick_clean_df = pd.read_csv('Resources/KickstarterData.csv')

# Read our Kickstarter data into pandas
kick_clean_df.head()
```

```python
# Get a list of all of our columns for easy reference
print(list(kick_clean_df.columns))
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
kick_clean_extracted_df = kick_clean_df[['name', 'goal', 'pledged', 'state', 'country',
                                         'staff_pick', 'backers_count', 'spotlight']]
kick_clean_extracted_df.head()
```

```python
# Remove projects that made no money at all
kick_clean_extracted_df.pledged.value_counts().head()
kick_clean_no_money = kick_clean_extracted_df.loc[kick_clean_extracted_df.pledged > 0]
kick_clean_no_money.head()
```

```python
# Collect only those projects that were hosted in the US
kick_clean_us_only = kick_clean_no_money.loc[kick_clean_no_money.country == 'US']
kick_clean_us_only.head()
```

```python
# Create a new column that finds the average amount pledged to a project
kick_clean_us_only['Avg donation'] = (kick_clean_us_only.pledged / kick_clean_us_only.backers_count)
kick_clean_us_only.head()
```
