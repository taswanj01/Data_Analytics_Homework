```python
X_train = games_train_filtered.drop("win_lose_draw", axis=1)
y_train = games_train_filtered["win_lose_draw"].values.reshape(-1, 1)
print(X_train.shape, y_train.shape)
```

```python
X_test = games_test_filtered.drop("win_lose_draw", axis=1)
y_test = games_test_filtered["win_lose_draw"].values.reshape(-1, 1)
print(X_test.shape, y_test.shape)
```

```python
# from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a logistic regression model

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_train, y_train)
print(f'Clf score on train data is {classifier.score(X_train, y_train)}')
print(f'Clf score on test data is {classifier.score(X_test, y_test)}')
```

```python
predictions = classifier.predict(X_test)
```

```python
accuracy = pd.DataFrame({"Prediction": list(predictions), "Actual": y_test.ravel()}).reset_index(drop=True)
accuracy['correct'] = accuracy.Prediction == accuracy.Actual
accuracy.head(15)
```

```python
y_actu = pd.Series(y_test.ravel(), name='Actual')
y_pred = pd.Series(predictions, name='Predicted')
df_confusion = pd.crosstab(y_actu, y_pred)
```

```python
df_confusion.columns = ['Win', 'Lose']
```

```python
df_confusion.rename(index={
    df_confusion.index[0]: 'Win',
    df_confusion.index[1]: 'Lose',
    df_confusion.index[2]: 'Draw'
}, inplace=True)
df_confusion
```
